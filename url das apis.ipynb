```python
import base64
import requests
import json

BASE_URL = "https://www.drogariasminasmais.com.br/_v/segment/graphql/v1?workspace=master&maxAge=short&appsEtag=remove&domain=store&locale=pt-BR&__bindingId=e59072cb-1f1f-4548-af34-928101d341d5&operationName=productSearchV3&variables=%7B%7D&extensions=%7B%22persistedQuery%22%3A%7B%22version%22%3A1%2C%22sha256Hash%22%3A%229177ba6f883473505dc99fcf2b679a6e270af6320a157f0798b92efeab98d5d3%22%2C%22sender%22%3A%22vtex.store-resources%400.x%22%2C%22provider%22%3A%22vtex.search-graphql%400.x%22%7D%2C%22variables%22%3A%22"
END_URL = "%3D%3D%22%7D"

TOTAL_PRODUCTS = 5826
CHUNK_SIZE = 85

total_products = []

# Loop para fazer a paginação
from_index = 0
while from_index < TOTAL_PRODUCTS:
    to_index = min(from_index + CHUNK_SIZE - 1, TOTAL_PRODUCTS - 1)
    
    # Criando o JSON de variáveis
    variables = {
        "hideUnavailableItems": True,
        "skusFilter": "ALL",
        "simulationBehavior": "default",
        "installmentCriteria": "MAX_WITHOUT_INTEREST",
        "productOriginVtex": False,
        "map": "c",
        "query": "medicamentos",
        "orderBy": "OrderByReleaseDateDESC",
        "from": from_index,
        "to": to_index,
        "selectedFacets": [{"key": "c", "value": "medicamentos"}],
        "operator": "and",
        "fuzzy": "0",
        "searchState": None,
        "facetsBehavior": "Static",
        "categoryTreeBehavior": "default",
        "withFacets": False,
        "advertisementOptions": {
            "showSponsored": True,
            "sponsoredCount": 3,
            "advertisementPlacement": "top_search",
            "repeatSponsoredProducts": True
        }
    }
    
    # Base64
    encoded_variables = base64.b64encode(json.dumps(variables).encode()).decode().rstrip("=")
    
    # Criando a URL final
    url = BASE_URL + encoded_variables + END_URL
    
    # Fazendo a requisição
    try:
        response = requests.get(url)
        if response.status_code == 200:
            try:
                data = response.json()
                
                # Verificando se a resposta tem o formato esperado
                if data and isinstance(data, dict):
                    products = data.get("data", {}).get("productSearch", {}).get("products", [])
                    if products:
                        total_products.extend(products)
                        print(f"Carregados {len(products)} produtos ({from_index} - {to_index})")
                    else:
                        print(f"Sem produtos encontrados para o intervalo ({from_index} - {to_index})")
                else:
                    print(f"Formato inesperado da resposta: {data}")
            except ValueError:
                print(f"Erro ao decodificar a resposta JSON ({from_index} - {to_index})")
        else:
            print(f"Erro na requisição ({from_index} - {to_index}): {response.status_code}")
    except Exception as e:
        print(f"Erro ao fazer a requisição ({from_index} - {to_index}): {e}")
    
    # Atualiza o índice para a próxima iteração
    from_index += CHUNK_SIZE

# Imprimir as informações de todos os produtos extraídos
for product in total_products:
    item = product.get("items", [])[0]  # Pegando o primeiro item
    ean = item.get("ean", "N/A")
    product_name = product.get("productName", "N/A")
    brand = product.get("brand", "N/A")
    description = product.get("description", "N/A")
    
    # Garantir que o preço seja um número válido (float) antes de tentar formatá-lo
    price = product.get("priceRange", {}).get("sellingPrice", {}).get("lowPrice", 0)
    list_price = product.get("priceRange", {}).get("listPrice", {}).get("lowPrice", 0)
    price = float(price) if price is not None else 0
    list_price = float(list_price) if list_price is not None else 0

    discount = list_price - price if list_price else 0

    print(f"Marca: {brand}")
    print(f"Nome do produto: {product_name}")
    print(f"EAN: {ean}")
    print(f"Descrição: {description}")
    print(f"Preço com desconto: R$ {price:.2f}")
    print(f"Preço sem desconto: R$ {list_price:.2f}")
    print(f"Valor do desconto: R$ {discount:.2f}")
    print("-" * 50)

print(f"Total de produtos extraídos: {len(total_products)}")

```

```python
import base64
import requests
import json
import csv

BASE_URL = "https://www.drogariasminasmais.com.br/_v/segment/graphql/v1?workspace=master&maxAge=short&appsEtag=remove&domain=store&locale=pt-BR&__bindingId=e59072cb-1f1f-4548-af34-928101d341d5&operationName=productSearchV3&variables=%7B%7D&extensions=%7B%22persistedQuery%22%3A%7B%22version%22%3A1%2C%22sha256Hash%22%3A%229177ba6f883473505dc99fcf2b679a6e270af6320a157f0798b92efeab98d5d3%22%2C%22sender%22%3A%22vtex.store-resources%400.x%22%2C%22provider%22%3A%22vtex.search-graphql%400.x%22%7D%2C%22variables%22%3A%22"
END_URL = "%3D%3D%22%7D"

TOTAL_PRODUCTS = 5826
CHUNK_SIZE = 85

total_products = []

# Loop para fazer a paginação
from_index = 0
while from_index < TOTAL_PRODUCTS:
    to_index = min(from_index + CHUNK_SIZE - 1, TOTAL_PRODUCTS - 1)
    
    # Criando o JSON de variáveis
    variables = {
        "hideUnavailableItems": True,
        "skusFilter": "ALL",
        "simulationBehavior": "default",
        "installmentCriteria": "MAX_WITHOUT_INTEREST",
        "productOriginVtex": False,
        "map": "c",
        "query": "medicamentos",
        "orderBy": "OrderByReleaseDateDESC",
        "from": from_index,
        "to": to_index,
        "selectedFacets": [{"key": "c", "value": "medicamentos"}],
        "operator": "and",
        "fuzzy": "0",
        "searchState": None,
        "facetsBehavior": "Static",
        "categoryTreeBehavior": "default",
        "withFacets": False,
        "advertisementOptions": {
            "showSponsored": True,
            "sponsoredCount": 3,
            "advertisementPlacement": "top_search",
            "repeatSponsoredProducts": True
        }
    }
    
    # Base64
    encoded_variables = base64.b64encode(json.dumps(variables).encode()).decode().rstrip("=")
    
    # Criando a URL final
    url = BASE_URL + encoded_variables + END_URL
    
    # Fazendo a requisição
    try:
        response = requests.get(url)
        if response.status_code == 200:
            try:
                data = response.json()
                
                # Verificando se a resposta tem o formato esperado
                if data and isinstance(data, dict):
                    products = data.get("data", {}).get("productSearch", {}).get("products", [])
                    if products:
                        total_products.extend(products)
                        print(f"Carregados {len(products)} produtos ({from_index} - {to_index})")
                    else:
                        print(f"Sem produtos encontrados para o intervalo ({from_index} - {to_index})")
                else:
                    print(f"Formato inesperado da resposta: {data}")
            except ValueError:
                print(f"Erro ao decodificar a resposta JSON ({from_index} - {to_index})")
        else:
            print(f"Erro na requisição ({from_index} - {to_index}): {response.status_code}")
    except Exception as e:
        print(f"Erro ao fazer a requisição ({from_index} - {to_index}): {e}")
    
    # Atualiza o índice para a próxima iteração
    from_index += CHUNK_SIZE

# Preparando os dados para escrever no CSV
products_data = []

for product in total_products:
    item = product.get("items", [])[0]  # Pegando o primeiro item
    ean = item.get("ean", "N/A")
    product_name = product.get("productName", "N/A")
    brand = product.get("brand", "N/A")
    
    # Garantir que o preço seja um número válido (float) antes de tentar formatá-lo
    price = product.get("priceRange", {}).get("sellingPrice", {}).get("lowPrice", 0)
    list_price = product.get("priceRange", {}).get("listPrice", {}).get("lowPrice", 0)
    price = float(price) if price is not None else 0
    list_price = float(list_price) if list_price is not None else 0

    # Calculando o valor do desconto
    discount = list_price - price if list_price else 0

    # Arredondando os valores de preço e desconto para duas casas decimais
    price = round(price, 2)
    list_price = round(list_price, 2)
    discount = round(discount, 2)

    # Adicionando os dados do produto à lista
    products_data.append([brand, product_name, ean, price, list_price, discount])


# Escrevendo os dados no arquivo CSV
with open('produtos.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Escrevendo o cabeçalho
    writer.writerow(["Marca", "Nome do Produto", "EAN", "Preço com Desconto", "Preço sem Desconto", "Valor do Desconto"])
    
    # Escrevendo os dados dos produtos
    writer.writerows(products_data)

print(f"Total de produtos extraídos e exportados para CSV: {len(total_products)}")

```
