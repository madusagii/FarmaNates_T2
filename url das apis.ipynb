```python
import json
import requests

# URL da API
url = "https://www.drogariasminasmais.com.br/_v/segment/graphql/v1?workspace=master&maxAge=short&appsEtag=remove&domain=store&locale=pt-BR&__bindingId=e59072cb-1f1f-4548-af34-928101d341d5&operationName=productSearchV3&variables=%7B%7D&extensions=%7B%22persistedQuery%22%3A%7B%22version%22%3A1%2C%22sha256Hash%22%3A%229177ba6f883473505dc99fcf2b679a6e270af6320a157f0798b92efeab98d5d3%22%2C%22sender%22%3A%22vtex.store-resources%400.x%22%2C%22provider%22%3A%22vtex.search-graphql%400.x%22%7D%2C%22variables%22%3A%22eyJoaWRlVW5hdmFpbGFibGVJdGVtcyI6dHJ1ZSwic2t1c0ZpbHRlciI6IkFMTCIsInNpbXVsYXRpb25CZWhhdmlvciI6ImRlZmF1bHQiLCJpbnN0YWxsbWVudENyaXRlcmlhIjoiTUFYX1dJVEhPVVRfSU5URVJFU1QiLCJwcm9kdWN0T3JpZ2luVnRleCI6ZmFsc2UsIm1hcCI6ImMiLCJxdWVyeSI6Im1lZGljYW1lbnRvcyIsIm9yZGVyQnkiOiJPcmRlckJ5UmVsZWFzZURhdGVERVNDIiwiZnJvbSI6MTUsInRvIjoyOSwic2VsZWN0ZWRGYWNldHMiOlt7ImtleSI6ImMiLCJ2YWx1ZSI6Im1lZGljYW1lbnRvcyJ9XSwib3BlcmF0b3IiOiJhbmQiLCJmdXp6eSI6IjAiLCJzZWFyY2hTdGF0ZSI6bnVsbCwiZmFjZXRzQmVoYXZpb3IiOiJTdGF0aWMiLCJjYXRlZ29yeVRyZWVCZWhhdmlvciI6ImRlZmF1bHQiLCJ3aXRoRmFjZXRzIjpmYWxzZSwiYWR2ZXJ0aXNlbWVudE9wdGlvbnMiOnsic2hvd1Nwb25zb3JlZCI6dHJ1ZSwic3BvbnNvcmVkQ291bnQiOjMsImFkdmVydGlzZW1lbnRQbGFjZW1lbnQiOiJ0b3Bfc2VhcmNoIiwicmVwZWF0U3BvbnNvcmVkUHJvZHVjdHMiOnRydWV9fQ%3D%3D%22%7D"

# Fazendo a requisição para a API
response = requests.get(url)

# Verificando se a resposta foi bem-sucedida
if response.status_code == 200:
    data = response.json()

    # Acessando a lista de produtos
    products = data.get("data", {}).get("productSearch", {}).get("products", [])

    # Verificando se há produtos na resposta
    if products:
        # Pegando os primeiros 15 produtos
        for product in products[:15]:
            # Extraindo informações
            product_name = product.get("productName", "N/A")
            brand = product.get("brand", "N/A")
            ean = product.get("ean", "N/A")
            descricao = product.get("description", "N/A")
            price = product.get("priceRange", {}).get("sellingPrice", {}).get("lowPrice", 0)
            list_price = product.get("priceRange", {}).get("listPrice", {}).get("lowPrice", 0)
            discount = list_price - price

            # Exibindo os resultados
            print(f"Marca: {brand}")
            print(f"Nome do produto: {product_name}")
            print(f"EAN: {ean}")
            print(f"Descricao: {descricao}")
            print(f"Preço com desconto: R$ {price:.2f}")
            print(f"Preço sem desconto: R$ {list_price:.2f}")
            print(f"Valor do desconto: R$ {discount:.2f}")
            print("-" * 50)
    else:
        print("Nenhum produto encontrado.")
else:
    print(f"Erro na requisição: {response.status_code}")
```
